```python
Working With Models
```

```python
6)Create the target data and feature data where target data is survived
```

```python
x = df.drop(['Survived','Name','Ticket'],axis = 1)
y = df.Survived
```

```python
x
```

```python
y
```

```python
7)apply label-encoding on categorical data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
enc = LabelEncoder()
```

```python
x['Sex'] = enc.fit_transform(x['Sex'])
```

```python
x['Cabin'] = enc.fit_transform(x['Cabin'])
x['Embarked'] = enc.fit_transform(x['Embarked'])
```

```python
x
```

```python
8)Split the data into Training and testing Set
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2)
```

```python
9) Create a Logistic regression model for Target and feature data
```

```python
from sklearn.linear_model import LogisticRegression 
```

```python
clf = LogisticRegression()
clf.fit(x_train,y_train)
```

```python
clf.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,confusion_matrix,f1_score
```

```python
y_pred = clf.predict(x_test)
```

```python
10) Display the Confusion Matrix
```

```python
confusion_matrix(y_pred,y_test)
```

```python
11) Find the Accuracy Score
```

```python
accuracy_score(y_pred,y_test)
```

12) Find the Precision Score

```python
precision_score(y_pred,y_test)
```

```python
13) Find the F1_Score
```

```python
f1_score(y_pred,y_test)
```

```python
14)Find the probability of testing data
```

```python
y_prob = clf.predict_proba(x_test)
```

```python
y_prob
```

```python
15)Display ROC Curve and find the AUC score
```

```python
from sklearn.metrics import roc_curve,roc_auc_score
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
y_pred_proba = clf.predict_proba(x_test)[::,1]
```

```python
y_pred_proba
```

```python
fpr , tpr , threshold = roc_curve(y_test , y_pred_proba)
```

16) Print FPR, TPR, and thresholds

```python
print("False Positive Rate (FPR):",fpr)
print("True Positive Rate (TPR):",tpr)
print("thresholds:",threshold)
```

```python
auc = roc_auc_score(y_test, y_pred_proba) 
```

```python
auc
```

```python
plt.plot(fpr,tpr,label="data 1, auc=" +str(auc))
plt.legend(loc=4)
plt.xlabel("False Positive Rate")
plt.ylabel("Positive Rate")
plt.show() 
```
